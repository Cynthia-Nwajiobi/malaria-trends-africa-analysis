```python
import pandas as pd

# Load the new dataset
df = pd.read_csv('C:/Users/Admin/Downloads/DatasetAfricaMalaria.csv')

# Preview the first few rows
df.head()
```

```python
# Confirm the unique countries
df['Country Name'].unique()
```

```python
# Rename relevant columns for clarity
df.rename(columns={
    'Country Name': 'Country',
    'Incidence of malaria (per 1,000 population at risk)': 'Malaria_Incidence',
    'Malaria cases reported': 'Malaria_Cases'
}, inplace=True)

# Preview
df[['Country', 'Year', 'Malaria_Incidence', 'Malaria_Cases']].head()
```

```python
# Size of the dataset
df.shape
```

```python
# Check for missing values in the key columns
df[['Malaria_Incidence', 'Malaria_Cases']].isnull().sum()
```

```python
# Drop the rows with missing values
df.dropna(subset=['Malaria_Incidence', 'Malaria_Cases'])
```

```python
# Data types
df.dtypes
```

```python
# Statistical Summary

df[['Malaria_Incidence', 'Malaria_Cases']].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Filter for years 2008–2017
df_10yrs = df[(df['Year'] >= 2008) & (df['Year'] <= 2017)]

# Group by country and sum total incidence
incidence_by_country = df_10yrs.groupby('Country')['Malaria_Incidence'].sum().reset_index()

# Top 5 countries with highest malaria incidence
top5 = incidence_by_country.sort_values(by='Malaria_Incidence', ascending=False).head(5)

# Bottom 5 countries with lowest incidence (non-zero)
bottom5 = incidence_by_country[incidence_by_country['Malaria_Incidence'] > 0] \
    .sort_values(by='Malaria_Incidence', ascending=True).head(5)

# Plotting
plt.figure(figsize=(12, 5))

# Top 5
plt.subplot(1, 2, 1)
sns.barplot(data=top5, x='Malaria_Incidence', y='Country', palette='Reds_r')
plt.title('Top 5 Countries with Highest Malaria Incidence (2008–2017)')
plt.xlabel('Total Incidence per 1,000 at Risk')
plt.ylabel('Country')

# Bottom 5
plt.subplot(1, 2, 2)
sns.barplot(data=bottom5, x='Malaria_Incidence', y='Country', palette='Greens')
plt.title('Top 5 Countries with Lowest Malaria Incidence (2008–2017)')
plt.xlabel('Total Incidence per 1,000 at Risk')
plt.ylabel('Country')

plt.tight_layout()
plt.show()
```

```python
# Group by country and sum total malaria cases for 2008–2017
cases_by_country = df[(df['Year'] >= 2008) & (df['Year'] <= 2017)]
cases_by_country = cases_by_country.groupby('Country')['Malaria_Cases'].sum().reset_index()

# Top 5 countries with highest reported cases
top5_cases = cases_by_country.sort_values(by='Malaria_Cases', ascending=False).head(5)

# Bottom 5 countries with lowest (non-zero) reported cases
bottom5_cases = cases_by_country[cases_by_country['Malaria_Cases'] > 0] \
    .sort_values(by='Malaria_Cases', ascending=True).head(5)

# Plotting both charts
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Top 5 reported cases
sns.barplot(data=top5_cases, x='Malaria_Cases', y='Country', ax=axes[0], palette='Blues_d')
axes[0].set_title('Top 5 Countries by Reported Malaria Cases (2008–2017)')
axes[0].set_xlabel('Total Reported Cases')
axes[0].set_ylabel('Country')

# Bottom 5 reported cases
sns.barplot(data=bottom5_cases, x='Malaria_Cases', y='Country', ax=axes[1], palette='Purples')
axes[1].set_title('Bottom 5 Countries by Reported Malaria Cases (2008–2017)')
axes[1].set_xlabel('Total Reported Cases')
axes[1].set_ylabel('Country')

plt.tight_layout()
plt.show()
```

```python
# Explore contributing factors to the malaria cases across countries

# Rename relevant columns
df.rename(columns={
    'Use of insecticide-treated bed nets (% of under-5 population)': 'Bed_Net_Usage',
    'Children with fever receiving antimalarial drugs (% of children under age 5 with fever)': 'Antimalarial_Treatment',
    'People using at least basic drinking water services (% of population)': 'Water_Access'

    }, inplace=True)

df.head()
```

```python
# We have way too many missing values to carry out a correlation testing. And these variables are too sensitive to manipulate. 
# N.B: There are only 6 missing values in Water_Access, so we can work with that variable.
```

```python
# Filter out missing values
df['Water_Access'].dropna()
```

```python
# Scatter plots to visualize potential relationship between reported malaria cases and water access
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='Water_Access',
    y='Malaria_Cases',
    hue='Country',
    legend=False,
    alpha=0.7
)

plt.title('Water Access vs Malaria Cases')
plt.xlabel('Access to Clean Drinking Water')
plt.ylabel('Reported Malaria Cases')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix (optional)
corr = df[['Malaria_Cases', 'Water_Access']].corr()
corr
```

```python
# No significant correlation between access to clean drinking water and reported malaria cases.
```
